# Quantum Number Generator
https://thequantuminsider.com/2019/11/04/quantum-programming-101-16-qubit-random-number-generator-tutorial/


```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
q = QuantumRegister(16, 'q')
c = ClassicalRegister(16, 'c')
circuit = QuantumCircuit(q, c)
```

```python
circuit.draw()
```

```python
circuit.h(q) # hadamard gate applied
circuit.measure(q, c) # store to classical bit
```

```python
# use quantum simulator: Aer
from qiskit.providers.aer import AerSimulator
sim = AerSimulator() # new sim object
```

```python
job = sim.run(circuit) # run experiment
result = job.result() # get results
result.get_counts() # return count dictionary
```
